```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df=pd.read_csv('C://Users//kp//Pictures//Assignments//Basic Statistics_Level 1//Q9_a.csv',header=0)
cars=pd.read_csv('C://Users//kp//Pictures//Assignments//Basic Statistics_Level 1//Q7.csv',header=0)
speed_values=df['speed'].value_counts()
```

```python
#Histogram
fig,ax=plt.subplots(1,1,figsize=(20,15))

ax.set_xlabel('Speed')
ax.set_ylabel('Frequency of the Speed Value')
ax.set_xticks(range(0,28,2))
ax.hist(df['speed'],bins=range(0,28,2),align='mid',color='green')
ax.set_title('Satisfying the Assignment Criteria')
ax.legend(labels='Speed Parameter Distribution')
plt.show()
```

```python
fig,bx=plt.subplots(1,1,figsize=(20,15))

bx.set_xlabel('Points,Scores')
#Setting X-Tick labels to text
bx.set_xticklabels(['Points','Scores'])
bx.set_ylabel('Scale of Values available for Points/Scores')
box=bx.boxplot([cars['Points'],cars['Score']],patch_artist=True)
#Plotting Multiple Boxplots in a single figure is enabled passing the values to be plotted as a list

colors = ['green', 'purple']
 
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
#Change box color in boxplot    

bx.set_title('Satisfying the Assignment Criteria-Part2')
bx.legend(labels=['Points','Scores'])
plt.show()
```
